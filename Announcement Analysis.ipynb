```python
#Load stopwords
stop = stopwords.words('english')

#load word lists
positive_words = pd.read_csv("LoughranMcDonald_Positive.csv")
litigious_words = pd.read_csv("LoughranMcDonald_Litigious.csv")
modal_strong_words = pd.read_csv("LoughranMcDonald_ModalStrong.csv")
modal_weak_words = pd.read_csv("LoughranMcDonald_ModalWeak.csv")
negative_words = pd.read_csv("LoughranMcDonald_Negative.csv")
uncertain_words = pd.read_csv("LoughranMcDonald_Uncertainty.csv")
capability_words = pd.read_csv("capability_words.csv")

wordlists = {}
wordlists['positive'] = positive_words['words'].apply(lambda x: x.lower())
wordlists['litigious'] = litigious_words['words'].apply(lambda x: x.lower())
wordlists['modal_strong'] = modal_strong_words['words'].apply(lambda x: x.lower())
wordlists['modal_weak'] = modal_weak_words['words'].apply(lambda x: x.lower())
wordlists['negative'] = negative_words['words'].apply(lambda x: x.lower())
wordlists['uncertain'] = uncertain_words['words'].apply(lambda x: x.lower())
wordlists['capability'] = capability_words['words'].apply(lambda x: x.lower())
```

```python
#load csv's
ann = pd.read_csv("static_announcement.csv")
comp = pd.read_csv("static_completion.csv")
ann_date = pd.read_csv("static_ann_date.csv")
ind = pd.read_csv("static_industry.csv")

#filter csv columns
important_columns_ann = ['Action ID', 'M&A Announcement', 'M&A Completion',
                         'Acq CAR P1', 'Acq CAR P2', 'Acq CAR P3', 
                         'Acq CAR P4', 'Acq CAR P5', 'Sel CAR P1', 
                         'Sel CAR P2', 'Sel CAR P3', 'Sel CAR P4', 
                         'Sel CAR P5', 'Announced Total Value (mil.)',
                         'Announce Date']
important_columns_comp = ['Action ID', 'Acq CAR P1', 'Acq CAR P2', 
                          'Acq CAR P3', 'Acq CAR P4', 'Acq CAR P5', 
                          'Sel CAR P1', 'Sel CAR P2', 'Sel CAR P3', 
                          'Sel CAR P4', 'Sel CAR P5']

ann_filter = ann[important_columns_ann]
comp_filter = comp[important_columns_comp]

#subset of table so that every entry has an announcement
ann_filter = ann_filter.dropna(subset=['M&A Announcement'])
ann_filter = ann_filter[ann_filter['M&A Announcement'] != 'na']
ann_filter = ann_filter[ann_filter['M&A Announcement'] != 'tbd']
ann_filter = ann_filter[ann_filter['M&A Announcement'] != 'tba']


#subset of table so that every entry has a completion
ann_no_comp = ann_filter.dropna(subset=['M&A Completion'])
ann_no_comp = ann_no_comp[ann_no_comp['M&A Completion'] != 'na']
ann_no_comp = ann_no_comp[ann_no_comp['M&A Completion'] != 'tbd']
ann_no_comp = ann_no_comp[ann_no_comp['M&A Completion'] != 'tba']
```

```python
#Helper functions

#CACX 69466926 Bloomberg
#69466926 normal
def find_from_ID(id):
    return ann[ann['Action ID'] == id]

#helper function to strip words of special characters and to remove numbers
def clean_word(word_to_clean):
    cleaned = ''.join(e for e in word_to_clean if e.isalnum())
    try:
        as_number = int(cleaned)
        return ''
    except:
        return cleaned
```

```python
#load faster by using stored data
loaded_pickle = True
try:
    #sampled by ID ann
    in_s = open(os.path.join("pickles", "word_per_id_ann"), 'rb')
    word_per_id_df_ann = pkl.load(in_s)
    in_s.close()
    
    #by word ann
    in_s = open(os.path.join("pickles", "word_occ_ann"), 'rb')
    word_occ_df_ann = pkl.load(in_s)
    in_s.close()

    #sampled by ID comp
    in_s = open(os.path.join("pickles", "word_per_id_comp"), 'rb')
    word_per_id_df_comp = pkl.load(in_s)
    in_s.close()
    
    #by word comp
    in_s = open(os.path.join("pickles", "word_occ_comp"), 'rb')
    word_occ_df_comp = pkl.load(in_s)
    in_s.close()
    
    in_s = open(os.path.join("pickles", "all_keywords"), 'rb')
    all_keywords = pkl.load(in_s)
    in_s.close()
except:
    loaded_pickle = False
```

```python
#Tokenize the text
if not loaded_pickle:    
    #create tokens from all texts (strip first because it's still headline)
    ann_words = [[clean_word(word) for word in document.decode("'iso-8859-1").lower().split() if clean_word(word) not in stop] 
                 for document in ann_filter['M&A Announcement']] 
    comp_words = [[clean_word(word) for word in document.decode("'iso-8859-1").lower().split() if clean_word(word) not in stop] 
                 for document in ann_no_comp['M&A Completion']] 
    ann_and_comp_words = ann_words + comp_words
    all_tokens = sum(ann_and_comp_words, [])

    #list of words that occur less than 15 times or are shorter than 4 characters
    tokens_once = set(clean_word(word) for word in set(all_tokens) if all_tokens.count(clean_word(word)) < 15 or len(word) <= 4)

    #all keywords for a given text that are not in the list compiled above
    filtered_announcements = [[clean_word(word) for word in text if clean_word(word) not in tokens_once]
             for text in ann_words]
    filtered_completions = [[clean_word(word) for word in text if clean_word(word) not in tokens_once]
             for text in comp_words]

    #all keywords for a given text with corresponding count
    list_of_counters_ann = [Counter(wordlist) for wordlist in filtered_announcements]
    list_of_counters_comp = [Counter(wordlist) for wordlist in filtered_completions]

    #a set of all keywords (used to create occurence table)
    all_keywords = set(all_tokens) - set(tokens_once)

    #list of all IDs for announcements and completions
    all_action_ids_ann = ann_filter['Action ID']
    all_action_ids_comp = ann_no_comp['Action ID']

    #helperscript to construct dfs below
    dict_of_word_freqs_ann = {}
    for num, aid in enumerate(all_action_ids_ann):
        dict_of_counts = {}
        for keyword in all_keywords:
            dict_of_counts[keyword] = list_of_counters_ann[num][keyword]
        dict_of_word_freqs_ann[aid] = dict_of_counts

    dict_of_word_freqs_comp = {}
    for num, aid in enumerate(all_action_ids_comp):
        dict_of_counts = {}
        for keyword in all_keywords:
            dict_of_counts[keyword] = list_of_counters_comp[num][keyword]
        dict_of_word_freqs_comp[aid] = dict_of_counts

    #sampled by ID ann
    word_per_id_df_ann = pd.DataFrame(dict_of_word_freqs_ann)
    #by word ann
    word_occ_df_ann = pd.DataFrame(dict_of_word_freqs_ann).T

    #sampled by ID comp
    word_per_id_df_comp = pd.DataFrame(dict_of_word_freqs_comp)
    #by word comp
    word_occ_df_comp = pd.DataFrame(dict_of_word_freqs_comp).T
```

```python
#get the number of keywords in an action id that are also in one of the McDonald word lists
def get_number_of_words_in_list(aid, which):
    aid_keywords = get_word_occurences_for_id_ann(aid).index.values
    counter_pos = 0
    for k in aid_keywords:
        if k in wordlists[which].values:
            counter_pos += 1
    return counter_pos
```

```python
#build dataframe for corr
all_action_ids_ann_fut = into_future_subset['Action ID']
print "Number of data points: %d" % len(all_action_ids_ann_fut)

dict_of_keyword_count = {}
for num, aid in enumerate(all_action_ids_ann_fut):
    #store all information in here
    dict_of_values = {}
    
    dict_of_values['different keywords'] = float(len(get_word_occurences_for_id_ann(aid)))
    dict_of_values['total keywords'] = float(sum(get_word_occurences_for_id_ann(aid)))
    
    dict_of_values['Acq CAR P1'] = float(get_info_for_id(aid)['Acq CAR P1'])
    dict_of_values['Acq CAR P1 abs'] = np.abs(float(get_info_for_id(aid)['Acq CAR P1']))
    dict_of_values['Acq CAR P2'] = float(get_info_for_id(aid)['Acq CAR P2'])
    dict_of_values['Acq CAR P2 abs'] = np.abs(float(get_info_for_id(aid)['Acq CAR P2']))
    dict_of_values['Acq CAR P3'] = float(get_info_for_id(aid)['Acq CAR P3'])
    dict_of_values['Acq CAR P3 abs'] = np.abs(float(get_info_for_id(aid)['Acq CAR P3']))
    dict_of_values['Acq CAR P4'] = float(get_info_for_id(aid)['Acq CAR P4'])
    dict_of_values['Acq CAR P4 abs'] = np.abs(float(get_info_for_id(aid)['Acq CAR P4']))
    dict_of_values['Acq CAR P5'] = float(get_info_for_id(aid)['Acq CAR P5'])
    dict_of_values['Acq CAR P5 abs'] = np.abs(float(get_info_for_id(aid)['Acq CAR P5']))
    
    dict_of_values['Sel CAR P1'] = float(get_info_for_id(aid)['Sel CAR P1'])
    dict_of_values['Sel CAR P1 abs'] = np.abs(float(get_info_for_id(aid)['Sel CAR P1']))
    dict_of_values['Sel CAR P2'] = float(get_info_for_id(aid)['Sel CAR P2'])
    dict_of_values['Sel CAR P2 abs'] = np.abs(float(get_info_for_id(aid)['Sel CAR P2']))
    dict_of_values['Sel CAR P3'] = float(get_info_for_id(aid)['Sel CAR P3'])
    dict_of_values['Sel CAR P3 abs'] = np.abs(float(get_info_for_id(aid)['Sel CAR P3']))
    dict_of_values['Sel CAR P4'] = float(get_info_for_id(aid)['Sel CAR P4'])
    dict_of_values['Sel CAR P4 abs'] = np.abs(float(get_info_for_id(aid)['Sel CAR P4']))
    dict_of_values['Sel CAR P5'] = float(get_info_for_id(aid)['Sel CAR P5'])
    dict_of_values['Sel CAR P5 abs'] = np.abs(float(get_info_for_id(aid)['Sel CAR P5']))
    
    text_length = float(len(get_info_for_id(aid)['M&A Announcement'].split(' ')))
    number_sentences = float(len(get_info_for_id(aid)['M&A Announcement'].split('.')))
    number_detected_sentences = float(len(sent_detector.tokenize(get_info_for_id(aid)['M&A Announcement'].strip())))
    
    dict_of_values['number sentences'] = number_sentences
    dict_of_values['text length'] = text_length
    
    #check so you do not divide by zero
    if number_sentences == 0:
        dict_of_values['avg sentence length'] = 0
    else:
        dict_of_values['avg sentence length'] = text_length / number_sentences
        
    if number_detected_sentences == 0:
        dict_of_values['avg sentence length detected'] = 0
    else:
        dict_of_values['avg sentence length detected'] = text_length / number_detected_sentences
    
    positive_words_num = float(get_number_of_words_in_list(aid, 'positive'))
    negative_words_num = float(get_number_of_words_in_list(aid, 'negative'))
    uncertain_words_num = float(get_number_of_words_in_list(aid, 'uncertain'))
    strong_modal_num = float(get_number_of_words_in_list(aid, 'modal_strong'))
    weak_modal_num = float(get_number_of_words_in_list(aid, 'modal_weak'))
    litigious_num = float(get_number_of_words_in_list(aid, 'litigious'))
    capability_num = float(get_number_of_words_in_list(aid, 'capability'))
    
    dict_of_values['strong modal words'] = strong_modal_num
    dict_of_values['weak modal words'] = weak_modal_num
    dict_of_values['litigious words'] = litigious_num
    dict_of_values['uncertain words'] = uncertain_words_num   
    dict_of_values['positive words'] = positive_words_num
    dict_of_values['negative words'] = negative_words_num
    dict_of_values['capability words'] = capability_num
    if text_length == 0:
        dict_of_values['positive ratio'] = 0.0
        dict_of_values['negative ratio'] = 0.0
        dict_of_values['uncertain ratio'] = 0.0
        dict_of_values['strong modal ratio'] = 0.0
        dict_of_values['weak modal ratio'] = 0.0
        dict_of_values['litigious ratio'] = 0.0
        dict_of_values['capability ratio'] = 0.0
        
        
    else:
        dict_of_values['positive ratio'] = positive_words_num / text_length
        dict_of_values['negative ratio'] = negative_words_num / text_length
        dict_of_values['uncertain ratio'] = uncertain_words_num / text_length
        dict_of_values['strong modal ratio'] = strong_modal_num / text_length
        dict_of_values['weak modal ratio'] = weak_modal_num / text_length
        dict_of_values['litigious ratio'] = litigious_num / text_length
        dict_of_values['capability ratio'] = capability_num / text_length
        
    if positive_words_num == 0:
        dict_of_values['negativity'] = 1.0
    else:
        dict_of_values['negativity'] = negative_words_num / positive_words_num
        
    #negativity 2 from paper
    if positive_words_num == 0 and negative_words_num == 0:
        dict_of_values['negativity 2'] = 0.0
    else:
        dict_of_values['negativity 2'] = (negative_words_num - positive_words_num)/float(negative_words_num+positive_words_num)

    
    dict_of_keyword_count[aid] = dict_of_values 

#build dataframe and transpose it so that data is in columns and not rows
num_keyword_df = pd.DataFrame(dict_of_keyword_count).T
```

```python
#Regression function: takes one output and n input lists

#http://stackoverflow.com/questions/11479064/multivariate-linear-regression-in-python
def reg_m(y, x):
    ones = np.ones(len(x[0]))
    X = sm.add_constant(np.column_stack((x[0], ones)))
    for ele in x[1:]:
        X = sm.add_constant(np.column_stack((ele, X)))
    results = sm.OLS(y, X).fit()
    return results
```

```python
def format_date(x, how='month'):
    year = str(x[-4:])
    if how=='month':
        month = x.split('/')[0]
        if len(month) == 1:
            month = '0' + month
        return year+month
    return year
data_length_time['date'] = data_length_time['Announce Date'].apply(lambda x: format_date(x))
data_length_time['year'] = data_length_time['Announce Date'].apply(lambda x: format_date(x, how=''))
```

```python
data_length_grouped = data_length_time.groupby('date')
```

```python
def plot_text_over_time():
    times = []
    real_times = []
    lengths = []
    start = 0
    for i, group in data_length_grouped:
        times.append(start)
        real_times.append(i)
        start+=1
        lengths.append(np.mean(group['text length']))
    plt.plot(times, lengths, marker='o', color='r', ls='')
```
